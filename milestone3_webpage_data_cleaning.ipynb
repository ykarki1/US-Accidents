I'm going to import public holidays of United States from wikipedia page by webscraping using beautifulsoup.
Source: https://en.wikipedia.org/wiki/Federal_holidays_in_the_United_States


```python
# Importing necessary modules
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://en.wikipedia.org/wiki/Federal_holidays_in_the_United_States"
r = requests.get(url)

soup = BeautifulSoup(r.content, "html.parser")
```

```python
all_tables = soup.find_all("table")
len(all_tables)
```

```python
holidays = soup.find("table", class_="wikitable")

# Table parser
df = pd.read_html(str(holidays))[0]
df
```

```python
df.drop(columns="Details")
```
